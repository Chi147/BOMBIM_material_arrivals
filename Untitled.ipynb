```python
import pandas as pd
!pip install openpyxl


# Load your IFC export
df = pd.read_csv("Test_LakesideVW00_Arch_ifc.csv", low_memory=False)

# Pick relevant columns
cols = ["Category", "Name", "Material", "Length", "Area", "Volume"]
bom = df[cols]

# Group by element type + material + name
bom_summary = bom.groupby(["Category", "Name", "Material"], dropna=False).agg(
    Quantity=("Name", "count"),
    TotalLength=("Length", "sum"),
    TotalArea=("Area", "sum"),
    TotalVolume=("Volume", "sum")
).reset_index()

# Save to Excel/CSV
bom_summary.to_csv("BOM_summary.csv", index=False)
print(bom_summary.head(20))
```

```python
# Save Cleaned BOM to Excel (OpenBOM supports .xlsx)
bom_summary.to_excel("Cleaned_BOM.xlsx", index=False)
```
